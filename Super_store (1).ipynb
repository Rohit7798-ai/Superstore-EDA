```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Sample - Superstore.csv', encoding='latin-1')
```

```python
df.head()
```

```python
df.info()
```

```python
drop_col = ['Row ID', 'Order ID', 'Customer ID', 'Postal Code']
df.drop(drop_col, axis=1, inplace=True)
```

```python
df.describe()
```

**Average Sales is 229.8580**

**Average Quantity is 3.789574**

**Average Profit is 28.6568**

```python
df.head()
```

```python
df['Customer Name'].value_counts().sort_values(ascending=False).head(10)
```

```python
sns.barplot(x=df['Customer Name'].value_counts().sort_values(ascending=False).head(10).index, y=df['Customer Name'].value_counts().sort_values(ascending=False).head(10).values)
plt.xticks(rotation=90)
plt.title('Top 10 Customer')
plt.xlabel('Customer Name')
plt.ylabel('Count')
plt.show()
```

**William Brown leads as the top customer, with others like John Lee and Matt Abelman also making 30+ purchases.**

```python
Top_10_product_by_profit = df[['Product Name' , 'Profit']].sort_values(by='Profit', ascending=False).head(10)
Top_10_product_by_profit
```

```python
Top_10_product_by_profit.groupby('Product Name').size().plot(kind='barh',color=sns.palettes.mpl_palette('Dark2'))
plt.title('Top 10 Product by Profit')
plt.xlabel('Count')
plt.ylabel('Product Name')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

**Canon imageCLASS 2200 Advanced Copier leads in profit, far ahead of other products like HP and GBC binding systems.**

```python
Top_10_product_by_Region = df[['Region' , 'Profit']].sort_values(by='Profit', ascending=False).head(10)
Top_10_product_by_Region
```

```python
figsize = (12, 1.2 * len(Top_10_product_by_Region['Region'].unique()))
plt.figure(figsize=figsize)
plt.title('Top 10 Product by Region')
sns.violinplot(Top_10_product_by_Region, x='Profit', y='Region', hue='Region', inner='stick', palette='Dark2', legend=False)
sns.despine(top=True, right=True, bottom=True, left=True)
```

**The Central region shows the widest profit range for top products, indicating strong performance. In contrast, the South and West regions have limited or lower profit spreads.**

```python
Top_10_products_by_sales = df[['Product Name' , 'Sales']].sort_values(by='Sales', ascending=False).head(10)
Top_10_products_by_sales
```

```python
Top_10_products_by_sales['Sales'].plot(kind='line', figsize=(8, 4))
plt.title('Top 10 Products by Sales')
plt.gca().spines[['top', 'right']].set_visible(False)
```

**The top-selling product has significantly higher sales than the rest, while others show a sharp drop, highlighting a big sales gap among the top 10.**

```python
Top_10_Region_by_sales = df[['Region' , 'Sales']].sort_values(by='Sales', ascending=False).head(10)
Top_10_Region_by_sales
```

```python
Top_10_Region_by_sales['Sales'].plot(kind='hist', bins=20, title='Sales')
plt.title('Top 10 Region by Sales')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

**Most regions have sales clustered between 8,000 and 11,000, but a few regions stand out with much higher sales, showing clear performance differences.**

```python
result = df.groupby(['Category' , 'Sub-Category'])['Profit'].sum()
result
```

```python
result.plot(kind='barh', figsize=(10, 5))
plt.title('Total Profit by Category and Sub-Category')
plt.xlabel('Profit')
plt.ylabel('Category, Sub-Category')
plt.tight_layout()
plt.show()
```

**Technology products like Copiers and Phones bring in the highest profits, while Furniture—especially Tables—shows major losses, indicating a stark contrast in category performance.**

```python
result_1 = df.groupby(['Category' , 'Sub-Category'])['Sales'].sum()
result_1
```

```python
result_1.plot(kind='barh', figsize=(10, 5))
plt.title('Total Sales by Category and Sub-Category')
plt.xlabel('Sales')
plt.ylabel('Category, Sub-Category')
plt.tight_layout()
plt.show()
```

**Phones and Chairs top the sales chart, but not all high-sales items bring profit—like Tables, which have strong sales but massive losses, showing sales don't always mean success.**

```python
top_sale_by_mode = df.groupby(['Ship Mode' , 'Segment'])['Sales'].sum()
top_sale_by_mode
```

```python
sns.barplot(x=top_sale_by_mode.index.get_level_values(0), y=top_sale_by_mode.values, hue=top_sale_by_mode.index.get_level_values(1))
plt.title('Top Sale by Mode and Segment')
plt.xlabel('Ship Mode')
plt.ylabel('Sales')
plt.show()
```

**Standard Class dominates across all segments—especially among Consumers—making it the top-performing shipping mode by sales, while Same Day shows the lowest overall activity.**

```python
ro =  df.groupby(['Ship Mode' , 'Segment'])['Profit'].sum()
ro
```

```python
sns.barplot(x=ro.index.get_level_values(0), y=ro.values, hue=ro.index.get_level_values(1))
plt.title('Top Profit by Mode and Segment')
```

**Standard Class not only leads in sales but also generates the highest profit—especially from Consumers—while Same Day shipping yields the lowest profit across all segments.**

```python
ter = df.groupby(['Country' , 'City'])['Quantity'].sum().sort_values(ascending=False).head(10)
```

```python
sns.barplot(x=ter.index.get_level_values(0), y=ter.values, hue=ter.index.get_level_values(1))
plt.title('Top Quantity by Country and City')
plt
```

**New York City leads in quantity sold by a clear margin, followed by Los Angeles—while smaller cities like Springfield and San Diego contribute the least.**

```python
sales_by_state = df.groupby('State')['Sales'].sum().sort_values(ascending=False)
display(sales_by_state.head())
```

```python
plt.figure(figsize=(10, 10))
sales_by_state.plot(kind='barh')
plt.title('Total Sales by State')
plt.xlabel('Total Sales')
plt.ylabel('State')
plt.tight_layout()
plt.show()
```

**California leads all U.S. states by a large margin in total sales, followed by New York and Texas, while many smaller states contribute very little in comparison.**

```python
sns.pairplot(df, hue='Profit', vars=['Sales', 'Quantity', 'Discount'])
plt.show()
```

**The scatter plot matrix shows that high discounts often lead to lower profits, while moderate sales and quantities tend to cluster around neutral to slightly positive profit levels.**
